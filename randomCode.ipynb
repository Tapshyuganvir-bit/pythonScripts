```python
import csv
import json

# Input CSV file path
csv_file = r"C:\Users\TapshyuGanvir\OneDrive - Torana Inc\Documents\APIAutomation\Instances_API_Result\files\secretkey\NoSQLDatabase_Secret.csv"
# Output JSON file path
json_file = r"C:\Users\TapshyuGanvir\OneDrive - Torana Inc\Documents\APIAutomation\Instances_API_Result\files\secretkey\json\noSQL_output.json"

data = {
    "vaultType": "internal",
    "info": {
        "secret": "All_NoSQL_Client_Secret_and_Password_Public_Secret",
        "type": "Public",
        "description": "All_NoSQL_Client_Secret_and_Password_Public_Secret"
    },
    "config": {}
}

with open(csv_file, mode="r", newline="", encoding="utf-8-sig") as f:  # utf-8-sig removes BOM
    reader = csv.DictReader(f)

    # Clean header names (strip spaces, standardize)
    reader.fieldnames = [h.strip() for h in reader.fieldnames]
    print("Detected CSV headers:", reader.fieldnames)

    for row in reader:
        # Clean row keys
        row = {k.strip(): (v.strip() if v else "") for k, v in row.items()}

        base_key = row.get("Key")  # Safe access
        auth_type = row.get("Authentication", "")

        if not base_key:
            continue  # Skip if no Key column

        # Case 1: AzureAccountKey
        if auth_type == "AzureAccountKey" and row.get("AccountKey"):
            data["config"][f"{base_key}_AccountKey"] = row["AccountKey"]

        # Case 2: AzureServicePrincipal
        elif auth_type == "AzureServicePrincipal":
            if row.get("AzureTenant"):
                data["config"][f"{base_key}_AzureTenant"] = row["AzureTenant"]
            if row.get("OAuthClientId"):
                data["config"][f"{base_key}_OAuthClientId"] = row["OAuthClientId"]
            if row.get("OAuthClientSecret"):
                data["config"][f"{base_key}_OAuthClientSecret"] = row["OAuthClientSecret"]

        # Case 3: SCRAM or UsernameandPassword
        elif "SCRAM" in auth_type or "UsernameandPassword" in auth_type:
            if row.get("Password"):
                data["config"][f"{base_key}_Password"] = row["Password"]

with open(json_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)

print("‚úÖ JSON file created successfully:", json_file)
```

```python
import os
import sys
import time
import requests
import urllib3
import csv
from datetime import datetime

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

TOKEN_URL = 'https://qa.onprem.icedq.com/auth/realms/icedq.qa/protocol/openid-connect/token'
token_data = {
    'grant_type': 'password',
    'client_id': 'icedq.admin-ui',
    'client_secret': 'b2e25665-7fce-49fa-9c63-f7c872dabcf2',
    'username': 'tapshyu.g',
    'password': 'Admin@12345'
}
token_response = requests.post(TOKEN_URL, data=token_data, verify=False)
access_token = token_response.json().get("access_token")
if not access_token:
    print("‚ùå Failed to get token.")
    exit()
GET_SECRET_URL = 'https://qa.onprem.icedq.com/api/v1/secret/search?pageNo=1&pageSize=20&sort=updatedTimestamp:desc'
common_header = {
    'accept': 'application/json, text/plain, */*',
    'content-type': 'application/json',
    'org-id': 'org-icedq',
    'workspace-id': 'wksc-36655681-7ca9-58e1-b119-71ddb41cad1e',
    'authorization': f'Bearer {access_token}'
}
GET_SECRET_URL_PAYLOAD = {
    "updatedBetween": {
        "start": "10/28/2025 00:00:00 UTC",
        "end": "10/28/2025 23:59:59 UTC"
    }
}
search_secret_response = requests.post(GET_SECRET_URL, headers=common_header, json=GET_SECRET_URL_PAYLOAD, verify=False)
search_workflow_data = search_secret_response.json()

for item in search_workflow_data.get("items", []):
    secret_id = item["id"]
    delete_secret_url = f'https://qa.onprem.icedq.com/api/v1/secret/{secret_id}'
    print(delete_secret_url)
    response = requests.delete(delete_secret_url, headers=common_header,verify=False)

    print(f"Secret id {secret_id} is deleted: {response.status_code}")
    if response.status_code != 200:
        print("Error response:", response.text)
```

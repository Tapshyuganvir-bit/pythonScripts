```python
import pandas as pd
import json
from tabulate import tabulate 
import math
# Use a proper file path
file_path = r'C:\Users\TapshyuGanvir\Downloads\parquet_file\c-1.parquet'
df = pd.read_parquet(file_path, engine='pyarrow')

if 'column_info' not in df.columns or 'table_name' not in df.columns:
    print("Required columns 'column_info' or 'table_name' not found.")
else:
    try:
       
        user_input = input(f"Enter the row number to display (1 to {len(df)}): ")
        row_number = int(user_input) - 1  

        if 0 <= row_number < len(df):
            table_name = df.loc[row_number, 'table_name']
            json_str = df.loc[row_number, 'column_info']
            json_data = json.loads(json_str)
            table_df = pd.json_normalize(json_data)

            
            print(f"\nðŸ“„ Table for Row {row_number + 1} â€” Table Name: '{table_name}'")
            print(tabulate(table_df, headers='keys', tablefmt='fancy_grid', showindex=False))
        else:
            print("âŒ Invalid row number. Please enter a number within the valid range.")

    except ValueError:
        print("âŒ Please enter a valid integer.")
    except json.JSONDecodeError:
        print("âŒ The 'column_info' data is not valid JSON.")
    except Exception as e:
        print(f"âŒ An unexpected error occurred: {e}")
```

```python
import csv
import json

def csv_to_json():
    csv_file_path = r"C:\Users\TapshyuGanvir\Downloads\Data for demo\cmi-freshness_data.csv"       
    json_file_path = r"C:\Users\TapshyuGanvir\Downloads\Data for demo\cmi-freshness_data.json"   

    data_list = []

    with open(csv_file_path, mode='r', newline='') as csv_file:
        reader = csv.DictReader(csv_file)
        for row in reader:
            data_list.append({
                "metricValue": row["metric_value"],
                "metricDatetime": row["metric_timestamp"]
            })

    final_output = {"data": data_list}

    with open(json_file_path, 'w') as json_file:
        json.dump(final_output, json_file, indent=2)

    print(f"Conversion complete. JSON saved to: {json_file_path}")

csv_to_json()
```
